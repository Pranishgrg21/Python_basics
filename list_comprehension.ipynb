# List Comprehension in Python

## What are List Comprehensions?

A list comprehension is a compact way to create a list from an existing iterable (like another list, a range, a string, etc.). It combines a `for` loop, and optionally an `if` condition, into a single, readable line.

Let's compare creating a list of numbers with a `for` loop versus a list comprehension.

**Traditional `for` loop:**

```python
numbers = []
for i in range(1, 6):
    numbers.append(i)
print("Traditional for loop:", numbers)
```

**List comprehension**

```python
numbers_lc = [i for i in range(1, 6)]
print("List comprehension:", numbers_lc)
```

Both produce the same result, but the list comprehension is much cleaner\!

-----

## The Basic Syntax

The simplest form of a list comprehension follows this structure:

`new_list = [expression for item in iterable]`

  * **`expression`**: The operation you want to perform on each item (e.g., `item * 2`, `item.upper()`).
  * **`item`**: The variable representing each element in the iterable.
  * **`iterable`**: The sequence you are iterating over (e.g., `range(10)`, `my_list`, `a_string`).

### Example: Squaring Numbers

**Traditional `for` loop:**

```python
squares = []
squares
```

```python
for i in range(1, 6):
    print(i)
    squares.append(i ** 2)
```

```python
print(squares)
```

```python
squares_lc = [i ** 2 for i in range(1, 6)]
print(squares_lc)
```

```python
"""Example: Uppercasing words"""

words = ["apple", "banana", "cherry"]
```

```python
uppercase_words = []

for word in words:
    uppercase_words.append(word.upper())
print(uppercase_words)
```

```python
uppercase_lc = [word.upper() for word in words]
print(uppercase_lc)
```

# Adding Aa conditional Filter (`if`)

You can add an `if` condition to the end of the comprehension to filter items from the original iterable. Only items that satisfy the condition will be included in the new list.

The syntex is:
`new_list = [expression for item in iterable if condition]`

Example: Getting Only even numbers

Let's create a list of squares for only the even numbers between 0 and 9.

Traditional `for` loop

```python
even_square = []

for i in range(10):
    if i % 2 == 0:
        even_square.append(i)

print(even_square)
```

```python
even_square = []

for i in range(10):
    if i % 2 == 0:
        even_square.append(i * i)

print(even_square)
```

**List comprehension with `if`**

```python
even_square_lc = [i for i in range(20) if i % 2 == 0]
print(even_square_lc)
```

```python
even_square_lc = [i * i for i in range(20) if i % 2 == 0]
print(even_square_lc)
```

# Using Conditional Expression (`if/else`)

What if you want to apply a different expression based on a condition? You can use a full `if-else` conditional expression. The key difference is that the `if-else` caluse comes before the `for` loop.

The syntax is:
**`new_list = [expression_if_true if condition else
expression_if_false for items in iterable]`**

## Example: "Even and odd"

let's create a list that labels numbers from 0 to 9 as either "even" or "odd".

**Traditional `for` loop**

```python
labels = []
for i in range(1, 11):
    if i % 2 == 0:
        labels.append("even")
    else:
        labels.append("odd")

print(labels)
```

```python
even_numbers = []
odd_numbers = []

for i in range(1, 11):
    if i % 2 == 0:
        even_numbers.append(i)
    else: 
        odd_numbers.append(i)

print("Even_numbers:", even_numbers)
print("odd_numbers:", odd_numbers)
```

```python
even_numbers_lc = [i for i in range(1, 11) if i % 2 == 0]
odd_numbers_lc = [i for i in range(1, 11) if i % 2 != 0]

print("Even numbers:", even_numbers_lc)
print("odd numbers:", odd_numbers_lc)
```

```python
labels_lc = ["even" if i % 2 == 0 else "odd" for i in range(1, 11)]
print(labels_lc)
```

```python
grades = ["A", "B", "C", "D", "E", "F"]

["Pass" if grade == "A" else "Grade " + grade for grade in grades]
```

## Nested List Comprehensions

You can also nest list comprehensions, which is useful for working with nested lists (like matrices or 2D arrays).

## Example: Creating martix

```python
mar = []

for i in range(3):
    temp = []
    for j in range(1, 4):
        temp.append(j)
    mar.append(temp)

print(mar)
```

```python
mar_lc = [[j for j in range(1, 4)] for i in range(3)]
print(mar_lc)
```

### Exercise 4: Given a list of temperatures in Celsius, convert them to Fahrenheit.

The formula is: $F = (C * 9/5) + 32$

*Input:* `celsius_temps = [0, 10, 20, 30, 40]`  
*Expected Output:* `[32.0, 50.0, 68.0, 86.0, 104.0]`

```python
celsius_temps = [0, 10, 20, 30, 40]
```

```python
fahrenheit = []

for temp in celsius_temps:
    f = (temp * 9/5) + 32
    fahrenheit.append(f)
print(fahrenheit)
```

```python
fahrenheit_lc = [(temp * 9/5) + 32 for temp in celsius_temps]
print(fahrenheit_lc)
```

### Exercise 6: Filter Long Words. Given a sentence, create a list of all words that have more than 4 letters.

*Input:* `sentence = "the quick brown fox jumps over the lazy dog"`  
*Expected output:* `['quick', 'brown', 'jumps']`

```python
sentence = "the quick brown fox jumps over the lazy dog"
```

```python
list = []

words = sentence.split()

for word in words:
    if len(word) > 4:
        list.append(word)
print(list)
```

### Exercise 7: Negative and Positive. Given a list of numbers, create a new list where positive numbers remain the same and negative numbers are replaced by 0.

```python
numbers = [5, -1, 10, -8, 3, -4]
```

```python
result = []

for num in numbers:
    if num >= 0:
        result.append(num)
    else:
        result.append(0)
print(result)
```
